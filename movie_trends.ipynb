```python
def ratings(genre, year1, year2): 
    
    rating = []
    years = []
    
    for i in range(year1, year2):
        years.append(i)
    
    years.append(year2)
    
    for year in years:
        f1 = merge['year'] == str(year)
        f2 = merge['genres'].str.contains(genre)
        mean = merge[f1 & f2]['avg rating'].mean()
        rating.append(mean)

    return pd.DataFrame(rating, years, columns = [genre])
```

```python
def ratings1(genre, year1, year2): 
    
    rating = []
    
    for year in range(year1, year2+1):
        f1 = merge['year'] == str(year)
        f2 = merge['genres'].str.contains(genre)
        mean = merge[f1 & f2]['avg rating'].mean()
        rating.append(mean)

    return pd.DataFrame(rating, year, columns = [genre])
```
